```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
apollo_img = cv2.imread(os.path.relpath('./Data/Apollo_11_Launch.jpg'), 0)

plt.imshow(apollo_img, cmap="gray")
print(apollo_img)
```

```python
cropped_img = apollo_img[50:650, 400:800]
plt.imshow(cropped_img, cmap="gray")
```

```python
matrix = np.ones(cropped_img.shape, dtype="uint8") * 75

apollo_img_brighter = cv2.add(cropped_img, matrix)
apollo_img_darker   = cv2.subtract(cropped_img, matrix)

# Show the images
plt.figure(figsize=[18, 5])
plt.subplot(131); plt.imshow(apollo_img_darker, cmap="gray");  plt.title("Darker");
plt.subplot(132); plt.imshow(cropped_img, cmap="gray");         plt.title("Original");
plt.subplot(133); plt.imshow(apollo_img_brighter, cmap="gray");plt.title("Brighter");
```

```python
img_apollo_flipped_vert = cv2.flip(apollo_img_darker, 0)

# Show the images
plt.imshow(img_apollo_flipped_vert, cmap="gray")
```

```python
cv2.imwrite("apollo_img_edited.png", img_apollo_flipped_vert)
```
